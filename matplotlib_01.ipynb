```python
import matplotlib.pyplot as plt 
y=[23,45,72,47,27]
x=["part1","part2","part3","part4","part5"]
plt.bar(x,y,color="pink",edgecolor="red")
plt.xlabel("parts of harry poter ",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("popularity of different parts of harry poter",fontsize=15)
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
data=pd.read_excel("inter.xlsx")
df=pd.DataFrame(data)
print(df)
print()
grpby=df.groupby("Country")["salary"].sum()
print(grpby)
plt.bar(grpby.index,grpby.values,color="pink",edgecolor="red")
plt.xlabel("Country",fontsize="20")
plt.ylabel("Total Salary expense",fontsize="20")
plt.title("Total Expense for A Country",fontsize=20)
plt.show()

```
