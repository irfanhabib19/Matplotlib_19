```python
import matplotlib.pyplot as plt
x=[5,10,15,20,25]
y=[20,30,25,30,45]
y1=[23,45,43,56,20]
y3=[24,56,34,56,23]
y4=[20,40,65,35,37]
plt.figure(figsize=[6,4])
plt.plot(x,y,label="male")
plt.plot(x,y1,label="female")
plt.plot(x,y3,label="Kids")
plt.plot(x,y4,label="adult")
plt.legend(loc=0,ncols=2)
plt.show()
```
