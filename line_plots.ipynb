```python
import matplotlib.pyplot as plt
x=["Day1","Day2","Day3","Day4","Day5"]
y=[400,350,204,230,390]
y1=[300,450,270,243,220]
plt.plot(x,y,marker="o",color="red",ls="-",label="Week1",alpha=1)
plt.plot(x,y1,marker="^",color="blue",ls="-",label="Week2",alpha=0.5)
plt.legend(loc="upper right")

plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
data=pd.read_excel("expense3.xlsx")
df=pd.DataFrame(data)
#print(df)
gp=df.groupby("Category")["Amount"].sum()
print(gp)
plt.plot(gp.index,gp.values,color="red",alpha=1)
plt.show()
```
